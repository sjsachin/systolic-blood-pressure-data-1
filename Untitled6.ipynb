```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
data1=pd.read_csv("Ex03_SystolicBP_Regreesion.csv")
print(data1.info())
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
#print(data1.iloc[:,0])
#print(len(data1.columns))
```

```python
cor=data1.corr()["SystolicBP"][:-1]
#print(data1.corr()["SystolicBP"][:-1])
print(cor)
```

```python
plt.figure(figsize=(15,10.5))
plot_count = 1

for feature in list(data1.columns)[:-1]:
        plt.subplot(3,3,plot_count)
        plt.scatter(data1[feature], data1["SystolicBP"])
        plt.xlabel(feature.replace('_',' ').title())
        plt.ylabel('SystolicBP')
        plot_count+=1
plt.show() 
```

```python
#linear model for predicting Systolic BP
X1=data1.iloc[:,0:-1]
Y1=data1["SystolicBP"]
print(X1.info())
```

```python
#linear model for predicting Systolic BP
x_train_1,x_test_1,y_train_1,y_test_1=train_test_split(X1,Y1,test_size=0.4,random_state=4)
linreg=LinearRegression()
test_model=linreg.fit(x_train_1,y_train_1)
predicted_values_1=test_model.predict(x_test_1)
print(predicted_values_1)
```

```python
residual_1=predicted_values_1-y_test_1
#print(residual_1)
print("means of residuals")
print(residual_1.mean())
lst1=list()
lst2=list()
lst1=[100,105,110,115,120,125]
lst2=[-0.00876,-0.00876,-0.00876,-0.00876,-0.00876,-0.00876]
print("")
print("")
print("Residual Plot")
fig=plt.figure(figsize=(8,4))
plt.scatter(y_test_1,residual_1)
plt.plot(lst1,lst2,linestyle="dashed",color="black")
plt.xlabel("test_set(dependent_variable_value)",size=12)
plt.ylabel("residual_1",size=12)
plt.title("Residual Plot",size=12)
plt.show()

```

```python
print("intercept of the model")
print(test_model.intercept_)
```

```python
print("coefficient of independent variable")
print(test_model.coef_)
```

```python
print(r2_score(y_test_1,predicted_values_1))
```
